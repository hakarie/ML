```python
#Univariate Data: mark of 10 students
data=[45, 61, 90, 81, 60, 50, 90, 98, 83, 72]
n = len(data)

#sorted data
sorted_data = sorted(data)

#Mean
total  = 0
for val in data:
    total += val
mean = total/n

#Median
if n % 2 == 0:
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2])
else: 
    median  = sorted_data[n // 2]

#Mode
frequency = {}
for val in data:
    frequency[val] = frequency.get(val, 0) + 1
max_freq = max(frequency.values())
mode = []
if max_freq == 1:
    mode = none
else:
    for k, v in frequency.items():
        if(v == max_freq):
            mode.append(k)

# standard deviation
squared_diffs = [(x - mean) ** 2 for x in data]
variance = sum(squared_diffs) / n
std_dev = variance ** 0.5

#Mean Absolute Deviation (MAD)
mad = sum(abs(x - mean) for x in data) / n

#coefficient of variation (CV)
cv = (std_dev / mean) * 100

#skewness
m3 = sum((x - mean) ** 3 for x in data) / n
skewness = m3 / (std_dev ** 3)

#Kurtosis
m4 = sum((x- mean) ** 4 for x in data) / n
kurtosis = m4 / (std_dev ** 4)

print("Data: ", data)
print("Sorted Data: ", sorted_data)
print("Mean: ", round(mean, 2))
print("Median: ", median)
print("Mode: ", mode)
print("Standard Deviation: ", round(std_dev, 2))
print("Mean Absolute Deviation: ", round(mad, 2))
print("Coefficient of Variation: ", round(cv, 2), "%")
print("Skewness: ", round(skewness, 2))
print("Kurtosis (Excess): ", round(kurtosis, 2))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Create sample data
np.random.seed(0)
x = np.random.normal(loc=50, scale=10, size=100)
y = 2 * x + np.random.normal(loc=0, scale=15, size=100)

data = pd.DataFrame({'X': x, 'Y': y})

# Plot using seaborn scatterplot with regression line
sns.set(style="whitegrid")
plt.figure(figsize=(8,8))
sns.scatterplot(x='X', y='Y', data=data, color='blue', alpha=0.6)
sns.regplot(x='X', y='Y', data=data, scatter=False, color='red')

plt.title('Bivariate Scatter Plot with Regression Line')
plt.xlabel('X variable')
plt.ylabel('Y variable')
plt.show()
```
